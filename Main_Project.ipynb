## Selecting Sample for statistical analysis because our dataset is too big
### Let's select sample using Stratified Sampling technique

```python
# Determine the unique values of the stratification variable
strata = df['SEX'].unique()

# Initialize an empty DataFrame to store the sampled data
sampled_data = pd.DataFrame()

# Define the sample size as 5% of the total population size
total_population_size = len(df)
sample_size_per_stratum = int(0.05 * total_population_size)

# Perform stratified sampling
for stratum in strata:
    # Select data for the current stratum
    stratum_data = df[df['SEX'] == stratum]
    
    # Check if the stratum size is sufficient for sampling
    if len(stratum_data) >= sample_size_per_stratum:
        # take a random sample of size sample_size_per_stratum from the stratum
        sampled_stratum = stratum_data.sample(n=sample_size_per_stratum, random_state=42)
    else:
        # take the entire stratum as the sample
        sampled_stratum = stratum_data
    
    # Append the sampled stratum to the sampled data
    sampled_data = sampled_data.append(sampled_stratum)

# Reset index of the sampled data
sampled_data.reset_index(drop=True, inplace=True)
```

## Association between Age and Disease Conditions:

Null Hypothesis (H0): There is no association between age and the prevalence of disease conditions.

Alternative Hypothesis (H1): Older age groups are more likely to have certain disease conditions compared to younger age groups.

### Technique used : Chi-Square test

```python
# Define the age groups
age_groups = {
    'Children': (0, 18),
    'Young adults': (19, 35),
    'Middle-aged adults': (36, 60),
    'Older adults': (61, 94)
}

# Categorize age into different groups
sampled_data['AGE_GROUP'] = pd.cut(sampled_data['AGE'], 
                                    bins=[0, 18, 35, 60, 94], 
                                    labels=['Children', 'Young adults', 'Middle-aged adults', 'Older adults'])

# Display the first few rows of the dataframe with age groups
print(sampled_data[['AGE', 'AGE_GROUP']].head())
```

```python
# Create a contingency table of age groups and disease conditions
contingency_table = pd.crosstab(sampled_data['AGE_GROUP'], sampled_data['DIABETES'])

# Print the contingency table
print("Contingency Table:")
print(contingency_table)

# Perform the Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("\nChi-Square Test Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

## Result :

The Chi-Square test results indicate a significant association between age groups and the prevalence of diabetes. Older age groups show a higher likelihood of having diabetes compared to younger age groups. This conclusion is supported by a large Chi-Square test statistic (2346.46) and a very low p-value (approximately 0.0), suggesting strong evidence against the null hypothesis of no association. The expected frequencies table further confirms these findings, showing that the observed frequencies differ significantly from the expected frequencies under the null hypothesis.
